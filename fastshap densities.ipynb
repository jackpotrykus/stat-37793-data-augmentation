```python
import numpy as np
```

```python
fastshap_cutout = np.load('cutout_'+'shapley_values.npy')
fastshap_autoaug = np.load('autoaug_'+'shapley_values.npy')
fastshap_no_augs = np.load('no_augs_'+'shapley_values.npy')
```

```python
import pandas as pd
results_df = pd.DataFrame(data={"AutoAugment":fastshap_autoaug.reshape(-1), "Cutout":fastshap_cutout.reshape(-1),
                  "No Augmentations":fastshap_no_augs.reshape(-1)})
```

```python
results_df.plot.kde(title='Densities of Shapley Values')
```

```python
import matplotlib.pyplot as plt
results_df.plot.kde(title='Densities of Shapley Values: Right Tail')
plt.xlim(xmin=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
results_df.plot.kde(title='Densities of Shapley Values: Left Tail')
plt.xlim(xmax=0)
plt.show()
```
